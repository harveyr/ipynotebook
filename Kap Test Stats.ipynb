```python
from IPython.display import HTML, display
from string import Template

attempts = []
summaries = {}

student_a = "Student A"
student_b = "Student B"
student_c = "Student C"

dimensions = {
    'qbank': {
        'test_type': 'qbank',
        'questions': [1, 2, 3, 4]
    },
    'fixed': {
        'test_type': 'fixed',
        'questions': [5, 6, 7, 8]
    }
}


class TestStats:
    def __init__(self):
        self.answered = 0
        self.incorrect = 0
        self.seconds = 0

    def __str__(self):
        return '<TestStats: {}/{} ({} s)>'.format(
            self.correct,
            self.answered,
            self.seconds)

    def add(self, stats):
        self.answered += stats.answered
        self.incorrect += stats.incorrect
        self.seconds += stats.seconds

    @property
    def correct(self):
        return self.answered - self.incorrect

    @property
    def score(self):
        return int(100 * float(self.correct) / float(self.answered))

    @property
    def secondsPerQuestion(self):
        return self.seconds / self.answered


class TestAttempt(TestStats):

    def __init__(self, name, dimension, answers, seconds):
        self.name = name
        self.answers = answers
        self.answered = len(answers)
        self.incorrect = len([q for q in answers if not q])
        self.seconds = seconds
        self.dimension = dimension


def print_stat(key, val):
    tabs = '\t'
    if len(key) < 6:
        tabs += '\t' 
    print("|\t{}:{}{}".format(key, tabs, val))

def combined_stats(stats_list):
    combined = TestStats()
    combined.answered = sum([s.answered for s in stats_list])
    combined.incorrect = sum([s.incorrect for s in stats_list])
    combined.seconds = sum([s.seconds for s in stats_list])
    return combined

def global_stats():
    gs = TestStats()
    summarized = [v for k, v in summaries.iteritems()]
    gs.answered = sum([a.answered for a in summarized])
    gs.incorrect = sum([a.incorrect for a in summarized])
    gs.seconds = sum([a.seconds for a in summarized])
    return gs

def student_stats_for_type(name, test_type):
    atts = []
    for a in attempts:
        if a.name == name and dimensions[a.dimension]['test_type'] == test_type:
            atts.append(a)
    print('atts: {v}'.format(v=atts))
    return combined_stats(atts)

def global_stats_for_test_type(test_type):
    gs = TestStats()
    dims = []
    for dim_name in dimensions:
        if dimensions[dim_name]['test_type'] == test_type:
            dims.append(dim_name)
    atts = [a for a in attempts if a.dimension in dims]
    return combined_stats(atts)
    

def print_stats_vs_global(stats_list):
    gs = global_stats()
    row_template = Template("""
        <tr>
            <td> $label </td>
            <td> $answered </td>
            <td> $correct </td>
            <td> $score </td>
            <td style="$devStyle"><em>$scoreDev</em></td>
            <td> $seconds </td>
            <td> $secPerQuestion </td>
        </tr>
    """)

    html = """
    <table>
        <tr>
            <th> Stats </th>
            <th> Answered </th>
            <th> Correct </th>
            <th> Score </th>
            <th> <em>(Deviation)</em> </th>
            <th> Seconds </th>
            <th> Sec/Q </th>
        </tr>
    """

    gs = global_stats()
    html += row_template.substitute(dict(
        label="Global Stats",
            answered=gs.answered,
            correct=gs.correct,
            score="{}%".format(gs.score),
            devStyle="",
            scoreDev="-",
            seconds=gs.seconds,
            secPerQuestion=gs.secondsPerQuestion
    ))

    for stats_dict in stats_list:
        stats = stats_dict['stats']
        devStyle = "color: "
        scoreDev = stats.score - gs.score
        if scoreDev > 0:
            devStyle += "green"
        elif scoreDev < 0:
            devStyle += "red"
        html += row_template.substitute(dict(
            label=stats_dict['label'],
            answered=stats.answered,
            correct=stats.correct,
            score="{}%".format(stats.score),
            scoreDev="{}%".format(scoreDev),
            devStyle=devStyle,
            seconds=stats.seconds,
            secPerQuestion=stats.secondsPerQuestion

        ))
    html += "</table>"
    display(HTML(html))


def named_attempts(name):
    return [a for a in attempts if a.name == name]

def typed_attempts(test_type):
    return [a for a in attempts if a.test_type == test_type]


def print_stats(attempt):
    count = len(named_attempts(attempt.name))
    html = "<h4>{}&mdash;Attempt {}</h4>".format(attempt.name, count)
    display(HTML(html))
    student_aggregate_key = attempt.name + " Aggregate"
    student_attempt_key = "{} Attempt".format(attempt.name)
   
    test_type = dimensions[attempt.dimension]['test_type']
    student_type_stats = student_stats_for_type(attempt.name, test_type)
    global_type_stats = global_stats_for_test_type(test_type)

    print_stats_vs_global(
        [
            {
                'label': "{} Attempt".format(attempt.name),
                'stats': attempt
            },
            {
                'label': "{} Summary".format(attempt.name),
                'stats': summaries[attempt.name]
            },
            {
                'label': "{}'s {}s".format(attempt.name, test_type),
                'stats': student_type_stats
            },
            {
                'label': "Global {}s".format(test_type),
                'stats': global_type_stats
            }
        ]
    )


def update_summary(name):
    s = TestStats()
    latest_answers = {}
    seconds = 0
    for attempt in named_attempts(name):
        questions = dimensions[attempt.dimension]['questions']
        for question_idx in range(len(attempt.answers)):
            question_id = questions[question_idx]
            # We want the most current stats for each question (latest attempt)
            latest_answers[question_id] = attempt.answers[question_idx]
        seconds += attempt.seconds

    incorrect = len(
        [correct for key, correct in latest_answers.iteritems() if not correct])
    summary = TestStats()
    summary.answered = len(latest_answers)
    summary.incorrect = incorrect
    summary.seconds = seconds
    summaries[name] = summary

def add_attempt(name, dimension, answers, time):
    attempt = TestAttempt(name, dimension, answers, time)
    attempts.append(attempt)
    update_summary(name)
    print_stats(attempt)


add_attempt(student_a, 'qbank', [True, False, True, True], 180)
add_attempt(student_b, 'qbank', [True, True, True], 50)
add_attempt(student_c, 'qbank', [True, False], 30)
add_attempt(student_a, 'qbank', [False], 30)
add_attempt(student_a, 'fixed', [False, False, False, True], 4)
```
