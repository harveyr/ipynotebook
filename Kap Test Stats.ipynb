```python
from IPython.display import HTML
from IPython.display import display_html
```

```python

attempts = []
summaries = {}


class TestStats:
    def __init__(self):
        self.answered = 0
        self.incorrect = 0
        self.seconds = 0

    def add(self, stats):
        self.answered += stats.answered
        self.incorrect += stats.incorrect
        self.seconds += stats.seconds

    @property
    def correct(self):
        return self.answered - self.incorrect

    @property
    def score(self):
        return int(100 * float(self.correct) / float(self.answered))


class TestAttempt(TestStats):
    name = None
    test_type = 'qbank'

    def __init__(self, name, questions_correct, seconds):
        self.name = name
        self.questions_correct = questions_correct
        self.answered = len(questions_correct)
        self.incorrect = len([q for q in self.questions_correct if not q])
        self.seconds = seconds

def global_stats():
    gs = TestStats()
    summarized = [v for k, v in summaries.iteritems()]
    gs.answered = sum([a.answered for a in summarized])
    gs.incorrect = sum([a.incorrect for a in summarized])
    gs.seconds = sum([a.seconds for a in summarized])
    return gs


def print_stat(key, val):
    tabs = '\t'
    if len(key) < 6:
        tabs += '\t' 
    print("|\t{}:{}{}".format(key, tabs, val))


def print_stats_vs_global(stats_dict):
    gs = global_stats()
    html = """
    <table>
        <tr>
            <th> Stats </th>
            <th> Answered </th>
            <th> Correct </th>
            <th> Score </th>
            <th> Seconds </th>
        </tr>
    """

    for key in stats_dict:
        stats = stats_dict[key]
        html += """
        <tr>
            <td>{}</td>
            <td>{}</td>
            <td>{}%</td>
            <td>{}</td>
            <td>{}</td>
        </tr>
        """.format(
            key,
            stats.answered,
            stats.correct,
            stats.score,
            stats.seconds
        )
    html += "</table>"
    HTML(html)
    # print_stat('answered', stats.answered)
    # print_stat('incorrect', stats.incorrect)
    # print_stat('score', stats.score)
    # print_stat('(deviation)', stats.score - gs.score)
    # print_stat('time', stats.seconds)


def print_stats(attempt_no, attempt):
    # print("--- Attempt {} ({}) ---".format(attempt_no, attempt.name))
    student_aggregate_key = attempt.name + " Aggregate"
    student_attempt_key = "{} Attempt {}".format(attempt.name, attempt_no)
    print_stats_vs_global({
        student_aggregate_key: summaries[attempt.name],
        student_attempt_key: attempt
    })
    return
    print("Global:")
    # print_stat('answered', gs.answered)
    # print_stat('incorrect', gs.incorrect)
    # print_stat('score', gs.score)
    # print_stat('time', gs.seconds)

    print("")

    print("Student Aggregate:")
    # print_stats_vs_global(ag_stats)
    # print_stat('answered', ag_stats.answered)
    # print_stat('incorrect', ag_stats.incorrect)
    # print_stat('score', ag_stats.score)
    # print_stat('(deviation)', ag_stats.score - global_score)
    # print_stat('time', ag_stats.seconds)

    print("")

    print("Attempt {}".format(attempt_no))
    print_stats_vs_global(attempt)
    # print_stat('answered', attempt.answered)
    # print_stat('incorrect', attempt.incorrect)
    # print_stat('score', attempt.score)
    # print_stat('(deviation)', attempt.score - global_score)
    # print_stat('time', attempt.seconds)
    print("")


def named_attempts(name):
    return [a for a in attempts if a.name == name]


def update_summary(name):
    s = TestStats()
    questions_correct = {}
    seconds = 0
    for r in named_attempts(name):
        for i in range(len(r.questions_correct)):
            # We want the most current stats for each question (latest attempt)
            questions_correct[i] = r.questions_correct[i]
        seconds += r.seconds

    incorrect = len([correct for key, correct in questions_correct.iteritems() if not correct])
    summary = TestStats()
    summary.answered = len(questions_correct)
    summary.incorrect = incorrect
    summary.seconds = seconds
    summaries[name] = summary


def add_attempt(name, questions_correct, time):
    attempt = TestAttempt(name, questions_correct, time)
    attempts.append(attempt)
    update_summary(name)
    print_stats(len(attempts), attempt)


def print_test_type_stats(name):
    s = TestStats()
    for attempt in named_attempts(name):
        s.add(attempt)
```

```python
add_attempt('Student A', [True, False, True, True], 180)
add_attempt('Student B', [True, True, True], 50)
add_attempt('Student C', [True, False], 30)
add_attempt('Student A', [False], 30)
```
